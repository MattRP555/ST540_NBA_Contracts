# Merge Datasets

```python
import pandas as pd
import numpy as np
import re
```

### Load Full Dataset

```python
df_full = pd.read_csv('../data/full_nonrookie.csv')
df_full.head()
```

### Load Awards

```python
df_awds = pd.read_csv('../data/playerCumulativeAwards.csv')
df_awds.rename({'name': 'fullName', 'season': 'signSeason'}, axis=1, inplace=True)
df_awds.head()
```

### Merge

```python
df = df_full.merge(df_awds, on=['fullName', 'signSeason'], how='left').fillna(0)
df.head()
```

```python
cols = ['fullName', 'signSeason', 'totalValue', 'aav', 'nbaChamp', 'allStar',
        'allNBA1stTeam', 'allNBA2ndTeam', 'allNBA3rdTeam']
# df[df['fullName']=='Kevin Love'][cols]
# df[df['fullName']=='Paul George'][cols]
# df[df['fullName']=='Kawhi Leonard'][cols]
df[df['fullName']=='Chris Paul'][cols]
```

### Save dataset

```python
df.to_csv('../data/full_nonrookie_w_awards.csv', index=False)
```
